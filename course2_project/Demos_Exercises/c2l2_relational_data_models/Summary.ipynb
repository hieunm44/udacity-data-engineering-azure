## Demo 2: Create Denormalized Tables
Bắt đầu bằng 4 tables: `album_library, song_library, artist_library, song_length`.

<img src="images/table4.png" width=400> <img src="images/table5.png" width=400> <img src="images/table6.png" width=300> <img src="images/table7.png" width=300>

### Query 1: `select artist_name, album_name, year, song_name, song_length FROM <min number of tables>`
Ta thêm Artist Name vào table `album_library` và Song Length vào table `song_library`, tạo ra 2 tables `album_library2` và `song_library2`.

<img src="images/table9.png" width=400> <img src="images/table8.png" width=400>

Giờ chỉ cần 1 join để chạy query trên.

### Query 2: `select album_name SUM(song_length) FROM <min number of tables> GROUP BY album_name`
Ta ko muốn join, nên tạo table `album_length`.
