# Azure Data Pipeline Components

## 1. SQL Database
1. Tạo resource **SQL Database** -> Set như dưới, đặt DB name là `db_sales` -> **Next**
   <img src="images/azure1.png" width=800>

   Trong tab **Additional settings**, ở **Use existing data**, chọn **Sample**. Còn lại để nguyên -> **Review + create** -> **Create**

   <img src="images/azure2.png" width=1000>

2. Vào SQL Server vừa tạo (ko phải vào DB) -> **Security** -> **Networking** \
   Trong **Public access**, Chọn **Selected networks** -> **Add your client IPv4 address** -> Tick "Alow Azure services and resources to access this server" -> **Save**

   <img src="images/azure3.png" width=1500>

3. Vào DB đã tạo -> **Query editor** -> Đăng nhập để kiểm tra.
   ```sql
   SELECT TOP 10 *
   FROM [SalesLT].[Customer]
   ```

   <img src="images/azure4.png" width=1500>

## 2. Storage account
1. Tạo **Storage account** -> Set như dưới -> **Next**
   <img src="images/azure5.png" width=800>

   Trong tab **Advanced**, tick **Enable hierarchical namespace**. Còn lại để nguyên. -> **Review + create** -> **Create**

   <img src="images/azure6.png" width=800>

2. Vào Storage account vừa tạo tạo -> **Storage browser** -> **Blob containers** -> **Add container** -> Điền tên là `stagingcontainer` -> **Create**

   <img src="images/azure7.png" width=1500>

3. Vào container vừa tạo -> **Add directory** -> Điền tên  -> **Save**

   <img src="images/azure8.png" width=1500>

## 3. Synapse Analytics
1. Vào Synapse Studio ->  **Develop** -> **New SQL script** -> Chọn Dedicated pool đã tạo -> Chạy
    ```sql
    CREATE TABLE SalesOrderHeader(
    [SalesOrderID] [int] NOT NULL,
    [RevisionNumber] [tinyint] NOT NULL,
    [OrderDate] [datetime] NOT NULL,
    [DueDate] [datetime] NOT NULL,
    [ShipDate] [datetime] NULL,
    [Status] [tinyint] NOT NULL,
    [OnlineOrderFlag] [BIT] NOT NULL,
    [SalesOrderNumber] [nvarchar] (23) NOT null,
    [PurchaseOrderNumber] [nvarchar] (23) NULL,
    [AccountNumber] [nvarchar] (23) NULL,
    [CustomerID] [int] NOT NULL,
    [ShipToAddressID] [int] NULL,
    [BillToAddressID] [int] NULL,
    [ShipMethod] [nvarchar](50) NOT NULL,
    [CreditCardApprovalCode] [varchar](15) NULL,
    [SubTotal] [money] NOT NULL,
    [TaxAmt] [money] NOT NULL,
    [Freight] [money] NOT NULL,
    [TotalDue] [money] NULL,
    [Comment] [nvarchar] (1000) NULL,
    [rowguid] [uniqueidentifier] NOT NULL,
    [ModifiedDate] [datetime] NOT NULL
    )
    ```
2. Tạo 1 SQL sccript khác và chạy:
    ```sql
    CREATE TABLE Customer(
    [CustomerID] [int] NOT NULL,
    [NameStyle] [bit] NOT NULL,
    [Title] [nvarchar](8) NULL,
    [FirstName] [nvarchar] (100) NOT NULL,
    [MiddleName] [nvarchar](10) NULL,
    [LastName] [nvarchar] (100) NOT NULL,
    [Suffix] [nvarchar](10) NULL,
    [CompanyName] [nvarchar] (128) NULL,
    [SalesPerson] [nvarchar] (256) NULL,
    [EmailAddress] [nvarchar](50) NULL,
    [Phone] [nvarchar] (20) NULL,
    [PasswordHash] [varchar] (128) NOT NULL,
    [PasswordSalt] [varchar](10) NOT NULL,
    [rowguid] [uniqueidentifier] NOT NULL,
    [ModifiedDate] [datetime] NOT NULL
    )
    ```

## 4. Data Factory
1. Tạo resource **Data Factory** -> Set như dưới, còn lại để nguyên -> **Review + Create** -> **Create**

   <img src="images/azure9.png" width=800>

2. Vào Data Factory vừa tạo -> **Launch Studio** -> **Manage** -> Linked Service**, tạo link tới **Azure SQL Database** (Name là `ls_sql_sales`), **Lake Storage Gen 2** (Name là `ls_lake_storage`), và Synapse Analytics (Name là `ls_synapse`).

   <img src="images/azure10.png" width=500>
   <img src="images/azure11.png" width=500>
   <img src="images/azure12.png" width=500>

3. Vào **Author** -> **Datasets** -> **New dataset** -> **Azure SQL Database** \
   Đặt Name là `ds_sqldb_salesorderheader`, chọn Linked service vừa tạo, chọn Table `SalesLT.SalesOrderHeader` -> **OK** \
   Làm lại với Name là `ds_sqldb_customer` và chọn Table `SalesLT.Customer`.

   <img src="images/azure13.png" width=600>
   <img src="images/azure14.png" width=600>

   Tạo 1 Dataset nữa -> Chọn **Azure Synapse Analytics** -> Đặt Name là `ds_synapse_salesorderheader`, chọn Table `dbo.SalesOrderHeader` -> **OK** \
   Làm lại với Name là `ds_synapse_customer` và chọn Table `dbo.Customer`.
   <img src="images/azure15.png" width=600>
   <img src="images/azure16.png" width=600>

4. Ấn **Publish all** -> **Publish**

   <img src="images/azure17.png" width=1000>

## 4. Integration Runtime
Vào **Manage** -> **Integration runtimes** -> **New** -> **Azure**, **Self-Hosted** -> **Azure** -> Đặt Name là `IRWestUS`, chọn Region là `West US`. \
Trong tab **Data flow runtime**, chọn **Compute size** là `Custom`, chú ý phần **Time to live**, set như dưới -> **Create**

<img src="images/azure18.png" width=600>
